### Probamos graficos del indice de refacción en particulas en bulto y peliculas delgadas

Se grafican los datos de indice de refracción del oro
Datos de Jhonson Y cristy para las partículas de oro en bluto (1972)

https://refractiveindex.info/?shelf=main&book=Au&page=Johnson

y Datos de Rosenblatt para las partículas de oro en películas delgadas de 11 nm (2020)

https://refractiveindex.info/?shelf=main&book=Au&page=Rosenblatt-11nm

```python
#LIBRERIAS
from scipy.interpolate import InterpolatedUnivariateSpline     #Para interpolar
import numpy as np                                             #Funciones matematicas
import matplotlib.pyplot as plt                                #Manipulación de graficos
import pandas as pd                                            #Base de datos
import scipy as sp                                             #Funciones especiales
import math               
```

```python
### Exportamos el documento de una carpeta llamada función y ubicada en el escritorio y lo guardamos en un 'data frame'
data_frame = pd.read_csv("Jhonson.csv")     ##Datos de particulas en bulto
data_frame2 = pd.read_csv("thin.csv")       ##Datos peliculas delgadas

#Se dividen en arreglos cada columna

#--------BULK DATA-----------#
x = data_frame['long_onda'] #(Ya son listas que dependen de X)
n = data_frame['n']
k = data_frame['k']

#--------THIN DATA-----------#
xp = data_frame2['long_onda'] #(Ya son listas que dependen de X)
np = data_frame2['n']
kp = data_frame2['k']

plt.plot(x,n,'-',label = 'n Partículas en Bulto', color = 'red')      #Graficamos n
plt.plot(x,n,'x', color = 'red')
plt.plot(x, k,'-', label = 'k Partículas en Bulto', color = 'black')      #Graficamos k
plt.plot(x,k,'x', color = 'black')

plt.plot(xp,np,'-',label = 'n Películas Delgadas', color = 'blue')      #Graficamos n
plt.plot(xp,np,'x', color = 'blue')
plt.plot(xp, kp,'-', label = 'k Películas Delgadas', color = 'green')      #Graficamos k
plt.plot(xp,kp,'x', color = 'g')


#Caracteristicas del gráfico
plt.xlabel("Longitud de onda (nm)")
plt.ylabel("n, k")
plt.legend()
plt.grid()
v = [-20,2100,0,15]
plt.axis(v)
```
